```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def crawl(url):
    headers={
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
    }

    response =requests.get(url,headers=headers)
    soup=BeautifulSoup(response.text,'html.parser')

    product=[]
    price=[]
    ratings=[]

    p_name=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
    p_price=soup.find_all('span',class_="a-price-whole")
    p_rating=soup.find_all('span',class_="a-size-base s-underline-text")

    for i in p_name:
        product.append(i.text)

    for i in p_price:
        price.append(i.text)

    for i in p_rating:
        ratings.append(i.text)

    min_len=min(len(product),len(price),len(ratings))
    product=product[:min_len]
    price=price[:min_len]
    ratings=ratings[:min_len]

    df=pd.DataFrame({'Product':product,'Price':price,'Ratings':ratings,})
    return df
    
```

```python
df=crawl("https://www.amazon.in/s?k=mobile+phones+under+15000&crid=2XATADS7OZOX4&sprefix=mobile+phones+under+%2Caps%2C208&ref=nb_sb_ss_ts-doa-p_2_20")
```

```python
def image_crawler(url):
    head = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"}
    response = requests.get(url,headers = head)
    soup = BeautifulSoup(response.text,'html.parser')

    image = []
    im_data = soup.find_all('img',class_="s-image")

    for i in im_data:
        image.append(i.get('src'))
    return image
```

```python
img=image_crawler("https://www.amazon.in/s?k=mobile+phones+under+15000&crid=2XATADS7OZOX4&sprefix=mobile+phones+under+%2Caps%2C208&ref=nb_sb_ss_ts-doa-p_2_20")
```

```python
def crawler(url):
    head={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36'}
    response=requests.get(url,headers=head)
    soup=BeautifulSoup(response.text,'html.parser')
    name=[]
    watchtime=[]
    rating=[]
    year=[]
    n=soup.find_all('h3',class_='ipc-title__text')
    p=soup.find_all('span',class_='sc-5bc66c50-6 OOdsw dli-title-metadata-item')
    r=soup.find_all('span',class_="ipc-rating-star--rating")
    y=soup.find_all('span',class_='sc-5bc66c50-6 OOdsw dli-title-metadata-item')
    for i in n:
        name.append(i.text)
    for i in p:
        text = i.text.strip()
        if "h" in text or "m" in text:
            watchtime.append(text)
        elif text.isdigit():
            year.append(text)
    for i in r:
        rating.append(i.text)
    for i in y:
        year.append(i.text)
    min_length = min(len(name), len(watchtime), len(rating),len(year))
    name = name[:min_length]
    year = year[:min_length]
    watchtime=watchtime[:min_length]
    rating = rating[:min_length]
    df=pd.DataFrame({'Name':name,'WatchTime':watchtime,'Rating':rating,'Year':year})
    return df
```

```python
df=crawler('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
df.to_csv('imdb.csv',index=False)
```
